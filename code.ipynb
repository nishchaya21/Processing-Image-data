Resizing the image using pillow library

```python
from PIL import Image
```

```python
img = Image.open('puppy.jpg')
img_resize = img.resize((200,200))
```

```python
img_resize.save('puppy_resized.jpg')
```

```python
# display the image
img = mpimg.imread('puppy_resized.jpg')
img_plot = plt.imshow(img)
plt.show()
```

```python
img.shape
```

Convert RGB to GreyScale using OpenCV

```python
import cv2
```

```python
img = cv2.imread('puppy_resized.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(img_grey)
```

```python
img_grey.shape
```

```python
cv2.imwrite('puppy_grey.jpg', img_grey)
```
